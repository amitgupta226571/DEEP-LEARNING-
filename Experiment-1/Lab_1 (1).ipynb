Create 1D, 2D, and 3D tensors using PyTorch and NumPy.
Show Basic Operations: Element-wise addition, subtraction, multiplication, and division.
Dot product and matrix multiplication.
Indexing and Slicing examples (Boolean masking, extracting subtensors)
Use .view(), .reshape(), .unsqueeze(), and .squeeze() in PyTorch. Compare with .reshape in Numpy
Broadcasting- Perform operations with tensors of different shapes.
In-place vs Out-of-place operations



```python
import numpy as np

# 1D tensor (vector)
a_np = np.array([1, 2, 3])

# 2D tensor (matrix)
b_np = np.array([[1, 2],
                 [3, 4]])

# 3D tensor
c_np = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])

print(a_np.shape)
print(b_np.shape)
print(c_np.shape)
```

```python
import torch

# 1D tensor
a_t = torch.tensor([1, 2, 3])

# 2D tensor
b_t = torch.tensor([[1, 2],
                    [3, 4]])

# 3D tensor
c_t = torch.tensor([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])

print(a_t.shape)
print(b_t.shape)
print(c_t.shape)
```

```python
x = torch.tensor([10, 20, 30])
y = torch.tensor([1, 2, 3])

print(x + y)
print(x - y)
print(x * y)
print(x / y)
```

```python
a = torch.tensor([1, 2, 3])
b = torch.tensor([4, 5, 6])

dot = torch.dot(a, b)
print(dot)  # 32
```

```python
np.dot(a_np, a_np)
```

```python
A = torch.tensor([[1, 2],
                  [3, 4]])

B = torch.tensor([[5, 6],
                  [7, 8]])

print(torch.matmul(A, B))
# OR
print(A @ B)
```

```python
np.matmul(b_np, b_np)
# OR
b_np @ b_np
```

```python
t = torch.tensor([[10, 20, 30],
                  [40, 50, 60]])

print(t[0, 1])
print(t[:, 1])
print(t[1, :])
```

```python
x = torch.tensor([1, 2, 3, 4, 5])

mask = x > 3
print(x[mask])  # tensor([4, 5])
```

```python
x_np = np.array([1, 2, 3, 4, 5])
print(x_np[x_np > 3])
```

```python
t = torch.arange(1, 13).reshape(3, 4)

sub = t[:2, 1:3]
print(sub)
```

```python
x = torch.arange(1, 7)

y = x.view(2, 3)
print(y)
```

```python
z = x.reshape(3, 2)
print(z)
```

```python
x = torch.tensor([1, 2, 3])

print(x.shape)           # [3]
print(x.unsqueeze(0))    # [1, 3]
print(x.unsqueeze(1))    # [3, 1]
```

```python
x = torch.tensor([[1, 2, 3]])

print(x.shape)        # [1, 3]
print(x.squeeze())    # [3]
```

```python
np.reshape(a_np, (3, 1))
np.expand_dims(a_np, axis=0)
np.squeeze(np.array([[1, 2, 3]]))
```

```python
A = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

B = torch.tensor([10, 20, 30])

print(A + B)
```

```python
x = torch.tensor([1, 2, 3])
y = x + 1

print(x)  # unchanged
```

```python
x = torch.tensor([1, 2, 3])
x.add_(1)

print(x)  # modified
```
